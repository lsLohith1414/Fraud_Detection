# Merging The Tables Based on Primary Keys, Forign Keys

## üß© Step 1 ‚Äî Start with BASE TABLE

```python
df = transaction_records.copy()
```

```python
df.head(2)
```

## üß© Step 2 ‚Äî Merge TARGET FIRST (mandatory)

```python
fraud_indicators.head(2)
```

```python
df = df.merge(fraud_indicators, on = "TransactionID", how = "left")
```

```python
df.head(2)
```

## üß© Step 3 ‚Äî Merge transaction-level tables

```python
transaction_metadata.head(2)
```

```python
df = df.merge(transaction_metadata, on = "TransactionID", how = "left")
```

```python
df.head(2)
```

```python
amount_data.head(2)
```

```python
df = df.merge(amount_data, on="TransactionID", how="left")
```

```python
df.head(2)
```

```python
anomaly_scores.head(2)
```

```python
df = df.merge(anomaly_scores, on="TransactionID", how="left")
```

```python
df.head(2)
```

```python
category_labels.head(2)
```

```python
df = df.merge(category_labels, on="TransactionID", how="left")
```

```python
df.head(2)
```

## üß© Step 4 ‚Äî Merge merchant data (via MerchantID)

```python
merchant_data.head(2)
```

```python
df = df.merge(
    merchant_data,
    on="MerchantID",
    how="left"
)
```

```python
df.head(2)
```

## üß© Step 5 ‚Äî Merge customer data (via CustomerID)

```python
customer_data.head(2)
```

```python
df = df.merge(
    customer_data,
    on="CustomerID",
    how="left"
)
```

```python
df.head(2)
```

## üß© Step 6 ‚Äî Merge customer behavior tables

```python
account_activity.head(2)
```

```python
df = df.merge(account_activity, on="CustomerID", how="left")
```

```python
df.head(2)
```

```python
suspicious_activity.head(2)
```

```python
df = df.merge(suspicious_activity, on="CustomerID", how="left")
```

```python
df.head(2)
```

# STEP 1 ‚Äî Post-Merge Validation (MANDATORY), Validation check, if tables are correctly merge or not 

## `If any assert fails ‚Üí STOP and fix merge.`

## ‚úî Row count must NOT increase

```python
assert df.shape[0] == transaction_records.shape[0], \
       "Row count changed after merge!"

df.shape[0] == transaction_records.shape[0]
```

## ‚úî TransactionID must stay unique

```python
assert df["TransactionID"].is_unique, \
       "Duplicate TransactionIDs found!"

df["TransactionID"].is_unique
```
