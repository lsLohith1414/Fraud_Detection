## üîÅ `eda_df = df` vs `eda_df = df.copy()`

This distinction is **critical** for safe EDA and feature engineering.

---

## 1Ô∏è‚É£ `eda_df = df` ‚ùå (NO COPY)

```python
eda_df = df
```

### What really happens?

- ‚ùå No new DataFrame is created  
- ‚ùå `eda_df` and `df` point to the **same object in memory**

**Mental model:**  
Two variable names ‚Üí **one notebook**

---

### üî• Dangerous Example

```python
eda_df = df
eda_df.drop("AnomalyScore", axis=1, inplace=True)
```

### üí• Result

- `"AnomalyScore"` is removed from **`df` as well**
- This happens even if you didn‚Äôt intend it

üëâ You just modified your **source data**

---

## 2Ô∏è‚É£ `eda_df = df.copy()` ‚úÖ (SAFE)

```python
eda_df = df.copy()
```

### What happens?

- ‚úÖ A **new, independent DataFrame** is created  
- ‚úÖ Changes to `eda_df` **do NOT affect** `df`

**Mental model:**  
Original notebook ‚Üí **Xerox copy**

---

### ‚úÖ Safe Example

```python
eda_df = df.copy()
eda_df["log_amount"] = np.log1p(eda_df["TransactionAmount"])
```

### ‚úî Outcome

- `df` remains untouched  
- EDA experiments are isolated  
- You can freely explore, drop, transform

---

## üß† Why This Matters for **Your Fraud Project**

In fraud detection, different stages have different roles:

| Stage              | Purpose                    |
|--------------------|----------------------------|
| Merged `df`        | Source of truth            |
| EDA                | Try transformations        |
| Feature Engineering| Create new features        |
| Modeling            | Needs clean, stable data   |

### üîë Rule You Must Follow

- **Always use `df.copy()` for EDA**
- **Never experiment directly on the source DataFrame**

---

## üéØ Final Rule (Memorize This)

> If you plan to modify a DataFrame for analysis or experiments,  
> **always use `df.copy()` ‚Äî never direct assignment.**

This one habit prevents silent, dangerous bugs.


## üîπ `eda_df = df` vs `df.copy()` vs Deep Copy ‚Äî **The Exact Truth**

---

## üîπ 1Ô∏è‚É£ `eda_df = df` ‚ùå  
### This is **NOT** a shallow copy

```python
eda_df = df
```

### What it really is:

üëâ **Reference assignment (aliasing)**

- ‚ùå No copy is made at all  
- ‚ùå Both variables point to the **same object**  
- ‚ùå Changes in one affect the other  

üìå **This is neither shallow nor deep copy**

---

## üîπ 2Ô∏è‚É£ `eda_df = df.copy()` ‚úÖ  
### This **IS** a shallow copy (by default)

```python
eda_df = df.copy()
```

### What happens:

- ‚úÖ Creates a **new DataFrame object**
- ‚úÖ Data is copied, but underlying objects *may* still be shared
- ‚úÖ Safe for **almost all EDA operations**

üìå In pandas terms:

> `df.copy()` = **shallow copy**

---

## üîπ 3Ô∏è‚É£ True Deep Copy (Rarely Needed)

```python
import copy
eda_df = copy.deepcopy(df)
```

### What this does:

- ‚úÖ Completely independent copy  
- ‚úÖ Copies all nested objects  
- ‚ùå Slower and memory-heavy  
- ‚ùå Almost never needed in EDA  

---

## üß† Pandas Reality (IMPORTANT)

| Code | Actual Meaning |
|-----|---------------|
| `eda_df = df` | Reference / alias ‚ùå |
| `df.copy()` | Shallow copy ‚úÖ |
| `copy.deepcopy(df)` | True deep copy |

---

## üî• Why This Confusion Happens

In daily practice, people casually say:

- ‚Äúdeep copy‚Äù for `df.copy()`
- ‚Äúshallow copy‚Äù for `eda_df = df`

But **technically**:

‚ùå That terminology is incorrect

üìå **Interviews may test this distinction**

---

## üß† Interview-Perfect Answer (USE THIS)

> `eda_df = df` does not create a copy; it creates a reference to the same DataFrame.  
> `df.copy()` creates a new DataFrame object (shallow copy), which is safe for EDA.  
> A true deep copy requires `copy.deepcopy()`, which is rarely needed.

---

## üéØ Practical Rule (MEMORIZE)

- **EDA ‚Üí `df.copy()`**
- **Never ‚Üí `eda_df = df` if modifying data**
- **Deep copy ‚Üí only if nested objects exist**

# Step-9: Bivariate Analysis

## A. Bivariate Analysis -> Numerical vs Categorical -> Input Numerical vs Target Categorical

````python
from scipy.stats import ttest_ind, f_oneway
````

````python
def bivariate_analysis_num_vs_cat(df,feature,target):

    # ---------------- Numerical vs Categorical ----------------
    print("Type: Numerical vs Categorical")
        
    groups = [df[df[target] == val][feature].dropna() for val in df[target].unique()]

        
    # t-test for binary target
    if len(groups) == 2:
        stat, p = ttest_ind(groups[0], groups[1])
        print(f"\nt-test: statistic={stat:.3f}, p-value={p:.4f}")
            
        
     # ANOVA for multi-class target
    elif len(groups) > 2:
        stat, p = f_oneway(*groups)
        print(f"\nANOVA: statistic={stat:.3f}, p-value={p:.4f}")



    # Visualization
    plt.style.use("Solarize_Light2")
    sns.set_theme(style="whitegrid")
    
    plt.figure(figsize = (15,6))
    plt.subplot(1,2,1)
    sns.boxplot(x=target, y=feature, hue = target, data=df ,palette="Set2")
    plt.title(f"{feature} vs {target}")

    plt.subplot(1,2,2)
    sns.violinplot(x=target, y=feature, data=df,hue = target , palette="Set2")
    plt.title(f"{feature} vs {target} (Distribution)")
    plt.show()
        
````

````python
for col in num_col:
    print(f"significant diffenrent analysis between {col} vs class")
    bivariate_analysis_num_vs_cat(eda_df,col,target)
    print("-"*140)
    print()
    print()
````

## B. Bivariate Analysis -> Categorical vs Categorical -> Input Categorical vs Target Categorical

````python
from scipy.stats import chi2_contingency
````

````python
def bivariate_analysis_cat_vs_cat(df,input_col,target_col ):
    crosstab = pd.crosstab(df[input_col],df[target_col])
    print(crosstab)
    chi2,p,dof,expected = chi2_contingency(crosstab)
    print(f"\nChi-Square Test: chi2={chi2:.3f}, p-value={p:.4f}")

    
    sns.countplot(x = input_col, hue=target_col, data = df )
    plt.title(f"{input_col} vs {target_col}")
    plt.show()
                                        
````

````python
cat_col.remove("FraudIndicator")
for col in cat_col:
    print(f"assosiation analysis between {col} vs class")
    bivariate_analysis_cat_vs_cat(eda_df,col,target)
    print("-"*140)
````

## üîë Short Answer (Very Important)

**YES** ‚Äî if a column fails a Chi-square test (or any univariate statistical test),  
you **CANNOT** conclude that the column is useless for prediction.

It only means:

> **‚ÄúThis column alone does not show a strong independent relationship with the target.‚Äù**

---

## üß† Why This Is TRUE (Core Reason)

Statistical tests like:

- Chi-square  
- t-test  
- ANOVA  
- Correlation  

are **UNIVARIATE tests**.

They answer **only one question**:

> **‚ÄúDoes this feature by itself have a statistically significant relationship with the target?‚Äù**

They **do NOT** capture:

- Interaction effects  
- Non-linear relationships  
- Conditional importance  
- Feature combinations  

---

## üîç Simple Example (Very Clear)

| Feature | Alone | Result |
|------|------|------|
| Feature A | Tested alone | ‚ùå Weak signal |
| Feature B | Tested alone | ‚ùå Weak signal |
| A + B together | Combined | ‚úÖ Strong signal |

üëâ This is **extremely common**, especially in fraud detection.

---

## üß† Fraud Example (REALISTIC)

| Feature | Alone | Combined |
|------|------|------|
| `txn_hour` | weak | üî• strong with `amount` |
| `is_weekend` | weak | üî• strong with `merchant` |
| `location` | weak | üî• strong with `device / velocity` |

üìå **Fraud lives in interactions, not single columns.**

---

## üî• Why Chi-square Can FAIL but Feature Still Helps

### 1Ô∏è‚É£ Non-linear effects
- Chi-square assumes simple dependency
- Tree-based models (RF, XGBoost) capture:
  - Thresholds
  - Splits
  - Non-linear patterns

### 2Ô∏è‚É£ Conditional relevance
- Example:
  - Weekend fraud ‚Üë **only when amount is high**
- Univariate tests miss this.

### 3Ô∏è‚É£ Class imbalance
- Fraud datasets are highly imbalanced
- Statistical tests lose power when:
  - Fraud rate is very low
  - Samples are uneven

---

## üß† Correct Interpretation of Test Failure

‚ùå **Wrong conclusion**

> ‚ÄúThis column is useless.‚Äù

‚úÖ **Correct conclusion**

> ‚ÄúThis column does not show strong independent predictive power and should be evaluated in combination with other features.‚Äù

---

## üß† How Professionals REALLY Use Statistical Tests

| Stage | Role of Statistical Tests |
|------|---------------------------|
| EDA | Screening & intuition |
| Feature selection | Optional guidance |
| Modeling | Models decide usefulness |
| Final decision | Based on model performance |

üëâ **Models decide predictive value, not tests.**

---

## üî• What Actually Determines Feature Usefulness?

- Model coefficients  
- Feature importance  
- SHAP values  
- Cross-validation performance  
- Interaction effects  

‚ùå Not just p-values.

---

## üß† Interview-Ready Answer (MEMORIZE)

> A statistical test like Chi-square evaluates the independent relationship between a single feature and the target. Failing such a test does not mean the feature is useless for prediction, as it may still contribute through interactions or non-linear effects when combined with other variables in a machine learning model.

---

## ‚úÖ Final Truth Table (LOCK THIS IN)

| Scenario | Conclusion |
|------|------|
| Fails Chi-square | ‚ùå Not useless |
| Passes Chi-square | ‚ùå Not guaranteed useful |
| Helps model performance | ‚úÖ Useful |
| Improves CV score | ‚úÖ Keep it |

---

## üß† Golden Rule (ONE LINE)

> **Statistical tests judge isolation.  
Models judge collaboration.**

# Spliting Data in to train test split

````python
from sklearn.model_selection import train_test_split
````

````python
X = eda_df.drop(columns= ['FraudIndicator'])
y = eda_df['FraudIndicator']
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.2, shuffle = True, random_state=42)
````

# building pipline 

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
````

````python
perprocessor = ColumnTransformer(transformers=[
    ("standard_scaler", StandardScaler(), ['TransactionAmount', 'AnomalyScore', 'Age','AccountBalance', 'txn_hour', 'txn_weekday','txn_month', 'days_since_last_login'], ),
    ("ohe", OneHotEncoder(handle_unknown="ignore"), ["Category"]), 
],remainder="drop"
)
````

````python
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression

final_pipeline = Pipeline(steps=[
    ("preprocessor", perprocessor),   # encoding + scaling
    ("smote", SMOTE(random_state=42)), # imbalance handling
    ("model", LogisticRegression(max_iter=1000))
])
````

````python
final_pipeline.fit(X_train, y_train)
````

````python
y_pred = final_pipeline.predict(X_test)
````

````python
from sklearn.metrics import accuracy_score, classification_report, recall_score, precision_score
````

````python
accuracy_score(y_pred,y_test)
````

````python
print(classification_report(y_pred,y_test))
````

````python
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from sklearn.linear_model import LogisticRegression
````

````python
cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)
````

````python
lr_params = {
    "model__penalty": ['l1', 'l2', 'elasticnet'],
    "model__C": [1,4,8],
    "model__class_weight":["balanced",None]
}
````

````python
rand_lr = RandomizedSearchCV(
    estimator=final_pipeline,
    param_distributions=lr_params,
    n_iter=50,
    scoring="accuracy",
    cv = cv,
    n_jobs=-1,
    verbose=1,
    random_state =42
    
)

rand_lr.fit(X_train,y_train)
````

````python
y_r_perd = rand_lr.predict(X_test)
````

````python
accuracy_score(y_r_perd,y_test)
````

````python
print(classification_report(y_r_perd,y_test))
````
