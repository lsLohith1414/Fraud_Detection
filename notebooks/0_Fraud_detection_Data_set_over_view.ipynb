# Step-9: Bivariate Analysis

## A. Bivariate Analysis -> Numerical vs Categorical -> Input Numerical vs Target Categorical

````python
from scipy.stats import ttest_ind, f_oneway
````

````python
def bivariate_analysis_num_vs_cat(df,feature,target):

    # ---------------- Numerical vs Categorical ----------------
    print("Type: Numerical vs Categorical")
        
    groups = [df[df[target] == val][feature].dropna() for val in df[target].unique()]

        
    # t-test for binary target
    if len(groups) == 2:
        stat, p = ttest_ind(groups[0], groups[1])
        print(f"\nt-test: statistic={stat:.3f}, p-value={p:.4f}")
            
        
     # ANOVA for multi-class target
    elif len(groups) > 2:
        stat, p = f_oneway(*groups)
        print(f"\nANOVA: statistic={stat:.3f}, p-value={p:.4f}")



    # Visualization
    plt.style.use("Solarize_Light2")
    sns.set_theme(style="whitegrid")
    
    plt.figure(figsize = (15,6))
    plt.subplot(1,2,1)
    sns.boxplot(x=target, y=feature, hue = target, data=df ,palette="Set2")
    plt.title(f"{feature} vs {target}")

    plt.subplot(1,2,2)
    sns.violinplot(x=target, y=feature, data=df,hue = target , palette="Set2")
    plt.title(f"{feature} vs {target} (Distribution)")
    plt.show()
        
````

````python
for col in num_col:
    print(f"significant diffenrent analysis between {col} vs class")
    bivariate_analysis_num_vs_cat(eda_df,col,target)
    print("-"*140)
    print()
    print()
````

## B. Bivariate Analysis -> Categorical vs Categorical -> Input Categorical vs Target Categorical

````python
from scipy.stats import chi2_contingency
````

````python
def bivariate_analysis_cat_vs_cat(df,input_col,target_col ):
    crosstab = pd.crosstab(df[input_col],df[target_col])
    print(crosstab)
    chi2,p,dof,expected = chi2_contingency(crosstab)
    print(f"\nChi-Square Test: chi2={chi2:.3f}, p-value={p:.4f}")

    
    sns.countplot(x = input_col, hue=target_col, data = df )
    plt.title(f"{input_col} vs {target_col}")
    plt.show()
                                        
````

## 6ï¸âƒ£ Transactions in the Past 1 Hour

````python
eda_df['Timestamp'] = pd.to_datetime(eda_df['Timestamp'])
eda_df = eda_df.sort_values(['CustomerID', 'Timestamp'])
````

````python
import numpy as np

# Convert timestamp
eda_df['Timestamp'] = pd.to_datetime(eda_df['Timestamp'], errors='coerce')
eda_df = eda_df.dropna(subset=['Timestamp'])

# Sort properly
eda_df = eda_df.sort_values(['CustomerID', 'Timestamp']).reset_index(drop=True)

# Empty column
eda_df['txns_last_1h'] = 0

# Loop per customer (FAST because each customer has few transactions)
for cid, group in eda_df.groupby('CustomerID'):
    times = group['Timestamp'].values
    result = np.zeros(len(times), dtype=int)
    
    for i in range(len(times)):
        # find how many timestamps fall within [current_time - 1h, current_time]
        cutoff = times[i] - np.timedelta64(1, 'h')
        result[i] = np.sum((times >= cutoff) & (times <= times[i]))
    
    eda_df.loc[group.index, 'txns_last_1h'] = result
````

## âœ… STEP 3 â€” CYCLIC ENCODING (on train, then test)

````python
import numpy as np

def cyc_encode(df):
    df['txn_hour_sin'] = np.sin(2*np.pi*df['txn_hour']/24)
    df['txn_hour_cos'] = np.cos(2*np.pi*df['txn_hour']/24)

    df['txn_weekday_sin'] = np.sin(2*np.pi*df['txn_weekday']/7)
    df['txn_weekday_cos'] = np.cos(2*np.pi*df['txn_weekday']/7)

    df['txn_month_sin'] = np.sin(2*np.pi*df['txn_month']/12)
    df['txn_month_cos'] = np.cos(2*np.pi*df['txn_month']/12)

    return df

X_train = cyc_encode(X_train)
X_test  = cyc_encode(X_test)

# drop raw time columns
time_cols = ['txn_hour', 'txn_weekday', 'txn_month']
X_train = X_train.drop(columns=time_cols)
X_test  = X_test.drop(columns=time_cols)
````

````python
# ======================================================
# 1. IMPORTS
# ======================================================
import mlflow
import mlflow.sklearn
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import json
import os

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.model_selection import StratifiedKFold, RandomizedSearchCV
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, average_precision_score,
    confusion_matrix,
    classification_report,
    precision_recall_curve,
    roc_curve
)
````

````python
# ======================================================
# 2. MLflow SETUP
# ======================================================
mlflow.set_tracking_uri("http://127.0.0.1:5000")
mlflow.set_experiment("FD-EXP1-OHE-O_S_ENN-EF26")

DATA_VERSION = "v1_ohe_smoteenn_fe26"
EXPERIMENT_DESCRIPTION = (
    "Experiment 1: Feature Engineering + OneHotEncoder + SMOTE-ENN + 32 engineered features"
)


# ======================================================
# 3. MODEL FACTORY
# ======================================================
model_factory = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(),
    "XGBoost": XGBClassifier(
        eval_metric="logloss",
        use_label_encoder=False
    ),
    "LightGBM": LGBMClassifier(verbose = -1),
}


# ======================================================
# 4. PARAM DISTRIBUTIONS (FIXED FOR SMOTE-ENN)
# ======================================================
param_grid = {

    "Logistic Regression": {
        "C": [0.001, 0.01, 0.1, 1, 10],
        "penalty": ["l1", "l2"],
        "solver": ["liblinear"],
        "class_weight": [None]
    },

    "Random Forest": {
        "n_estimators": [300, 500],
        "max_depth": [5, 10, None],
        "min_samples_split": [5, 10],
        "min_samples_leaf": [5, 10],
        "max_features": ["sqrt"],
        "class_weight": [None]
    },

    "XGBoost": {
        "n_estimators": [300, 500],
        "learning_rate": [0.01, 0.05],
        "max_depth": [3, 5],
        "min_child_weight": [5, 10],
        "subsample": [0.8],
        "colsample_bytree": [0.8],
        "gamma": [0, 0.5],
        "scale_pos_weight": [1]   # âœ… FIXED
    },

    "LightGBM": {
        "n_estimators": [300, 500],
        "learning_rate": [0.01, 0.05],
        "num_leaves": [31, 50, 100],
        "max_depth": [-1, 10],
        "min_child_samples": [20, 50],
        "subsample": [0.8],
        "colsample_bytree": [0.8],
        "reg_alpha": [0, 0.1],
        "reg_lambda": [0, 0.1],
        "class_weight": [None]
    }
}


# ======================================================
# 5. RANDOMIZED SEARCH FUNCTION
# ======================================================
def tune_model_random(model, param_distributions, X_train, y_train, n_iter=30):

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    search = RandomizedSearchCV(
        estimator=model,
        param_distributions=param_distributions,
        n_iter=n_iter,
        cv=cv,
        scoring="average_precision",
        n_jobs=-1,
        verbose=1,
        random_state=42,
        refit=True
    )

    search.fit(X_train, y_train)

    return search.best_estimator_, search.best_params_, search.best_score_


# ======================================================
# 6. NOTEBOOK EVALUATION FUNCTION
# ======================================================
def model_evaluation_notebook(model, X_test, y_test, model_name):

    y_pred = model.predict(X_test)

    if hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_test)[:, 1]
    else:
        y_proba = model.decision_function(X_test)

    print("\n" + "="*70)
    print(f"ðŸ“Š MODEL EVALUATION : {model_name}")
    print("="*70)

    print(f"Accuracy  : {accuracy_score(y_test, y_pred):.4f}")
    print(f"Precision : {precision_score(y_test, y_pred):.4f}")
    print(f"Recall    : {recall_score(y_test, y_pred):.4f}")
    print(f"F1 Score  : {f1_score(y_test, y_pred):.4f}")
    print(f"ROC-AUC   : {roc_auc_score(y_test, y_proba):.4f}")
    print(f"PR-AUC    : {average_precision_score(y_test, y_proba):.4f}")

    print("\nClassification Report:\n")
    print(classification_report(y_test, y_pred))

    cm = confusion_matrix(y_test, y_pred)

    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()

    precision, recall, thresholds = precision_recall_curve(y_test, y_proba)
    plt.figure(figsize=(6,4))
    plt.plot(recall, precision)
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.title("Precision-Recall Curve")
    plt.show()

    fpr, tpr, _ = roc_curve(y_test, y_proba)
    plt.figure(figsize=(6,4))
    plt.plot(fpr, tpr)
    plt.plot([0,1], [0,1], "--")
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.title("ROC Curve")
    plt.show()


# ======================================================
# 7. MASTER PIPELINE
# ======================================================
def run_full_pipeline(model_factory, param_grid, X_train, y_train, X_test, y_test):

    results = {}

    for name, model in model_factory.items():

        print(f"\nðŸ”· Training: {name}")

        with mlflow.start_run(run_name=name):

            mlflow.set_tag("data_version", DATA_VERSION)
            mlflow.set_tag("model_name", name)
            mlflow.set_tag("experiment_description", EXPERIMENT_DESCRIPTION)

            mlflow.log_param("n_rows", X_train.shape[0])
            mlflow.log_param("n_features", X_train.shape[1])
            mlflow.log_param("positive_ratio", float(np.mean(y_train)))

            os.makedirs("schema", exist_ok=True)
            schema_path = "schema/feature_schema.json"
            with open(schema_path, "w") as f:
                json.dump(list(X_train.columns), f, indent=4)
            mlflow.log_artifact(schema_path, artifact_path="schema")

            best_model, best_params, best_cv_score = tune_model_random(
                model,
                param_grid[name],
                X_train,
                y_train
            )

            for k, v in best_params.items():
                mlflow.log_param(k, v)

            mlflow.log_metric("best_cv_pr_auc", best_cv_score)

            y_pred = best_model.predict(X_test)

            if hasattr(best_model, "predict_proba"):
                y_proba = best_model.predict_proba(X_test)[:, 1]
            else:
                y_proba = best_model.decision_function(X_test)

            mlflow.log_metric("test_pr_auc", average_precision_score(y_test, y_proba))
            mlflow.log_metric("test_recall", recall_score(y_test, y_pred))
            mlflow.log_metric("test_precision", precision_score(y_test, y_pred))

            mlflow.sklearn.log_model(best_model, f"{name}_best_model")

            # ðŸ”¥ NOTEBOOK OUTPUT
            model_evaluation_notebook(best_model, X_test, y_test, name)

            results[name] = {
                "PR-AUC": average_precision_score(y_test, y_proba),
                "Recall": recall_score(y_test, y_pred),
                "Precision": precision_score(y_test, y_pred)
            }

    return results


# ======================================================
# 8. RUN PIPELINE
# ======================================================
results = run_full_pipeline(
    model_factory,
    param_grid,
    X_train_resampled_s_enn,
    y_train_resampled_s_enn,
    X_test,
    y_test
)

results
````

````python
from sklearn.preprocessing import RobustScaler, MinMaxScaler

ROBUST_FEATURES = [
    "Amount", "TransactionAmount", "AccountBalance",
    "user_avg_amount", "amount_last_24h",
    "merchant_avg_amount",
    "amount_diff_from_avg",
    "amount_diff_from_merchant_avg",
    "user_txn_count", "user_daily_txns",
    "txns_last_1h", "txns_last_24h",
    "merchant_txn_count",
    "days_since_last_login", "txn_gap_minutes"
]

MINMAX_FEATURES = [
    "AnomalyScore", "amount_ratio_avg", "Age"
]

def scale_train_test(
    X_train,
    X_test,
    robust_features,
    minmax_features
):
    """
    Fits scalers on TRAIN data and transforms both TRAIN and TEST.
    Simple, safe, production-ready.
    """

    X_train_scaled = X_train.copy()
    X_test_scaled = X_test.copy()

    # Robust scaling
    if robust_features:
        robust_scaler = RobustScaler()
        X_train_scaled[robust_features] = robust_scaler.fit_transform(
            X_train[robust_features]
        )
        X_test_scaled[robust_features] = robust_scaler.transform(
            X_test[robust_features]
        )

    # MinMax scaling
    if minmax_features:
        minmax_scaler = MinMaxScaler()
        X_train_scaled[minmax_features] = minmax_scaler.fit_transform(
            X_train[minmax_features]
        )
        X_test_scaled[minmax_features] = minmax_scaler.transform(
            X_test[minmax_features]
        )

    return X_train_scaled, X_test_scaled


X_train_scaled, X_test_scaled = scale_train_test(
    X_train,
    X_test,
    robust_features=ROBUST_FEATURES,
    minmax_features=MINMAX_FEATURES
)
````
