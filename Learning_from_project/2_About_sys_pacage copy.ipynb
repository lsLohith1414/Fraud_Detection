# sys.exc_info() â€” what is it?

`sys.exc_info()` is a function from the built-in **`sys`** module that gives you details about the exception that is **currently being handled**.

ðŸ“Œ It works **only inside an `except` block**.

---

## What does `sys.exc_info()` return?

It returns a **tuple of 3 items**:

```
(type, value, traceback)
```

### Meaning of each part

| Part | Meaning |
|-----|--------|
| `type` | The exception class (e.g. `ZeroDivisionError`) |
| `value` | The exception object / message |
| `traceback` | The traceback object (where the error happened) |

---

## Simple example (best way to understand)

```python
import sys

try:
    x = 10 / 0
except:
    exc_info = sys.exc_info()
    print(exc_info)
```

### Output (conceptually)

```
(<class 'ZeroDivisionError'>, ZeroDivisionError('division by zero'), <traceback object>)
```

---

## What is `exc_info`?

This line:

```python
exc_info = sys.exc_info()
```

means:

> â€œStore the current exception details in a variable named `exc_info`.â€

So:

```python
exc_info[0]  # exception type
exc_info[1]  # exception object / message
exc_info[2]  # traceback object
```

### Example

```python
print(exc_info[0].__name__)  # ZeroDivisionError
print(exc_info[1])          # division by zero
```

---

## Why do people use `sys.exc_info()` instead of `except Exception as e`?

### `except Exception as e` gives:
âœ” Exception object only

### `sys.exc_info()` gives:
âœ” Exception type  
âœ” Exception object  
âœ” Full traceback (very important)

Thatâ€™s why it is commonly used in:

- custom exception classes  
- logging systems  
- MLOps / production code  
- monitoring & observability pipelines  

---

## Very common real-world usage

```python
import sys
import traceback

try:
    x = 10 / 0
except:
    exc_type, exc_value, exc_tb = sys.exc_info()
    print(traceback.format_tb(exc_tb))
```

This prints the **exact line-by-line location** where the error occurred.
