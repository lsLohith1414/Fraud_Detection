```python
## 2. USER-LEVEL BEHAVIOR FEATURES

eda_df['user_avg_amount'] = eda_df.groupby('CustomerID')['TransactionAmount'].transform('mean')
eda_df['amount_diff_from_avg'] = eda_df['TransactionAmount'] - eda_df['user_avg_amount']
eda_df['amount_ratio_avg'] = eda_df['TransactionAmount'] / (eda_df['user_avg_amount'] + 1)
eda_df['user_txn_count'] = eda_df.groupby('CustomerID').cumcount() + 1
eda_df['date'] = pd.to_datetime(eda_df['Timestamp']).dt.date
eda_df['user_daily_txns'] = eda_df.groupby(['CustomerID', 'date']).cumcount() + 1

```

```python
## 3. VELOCITY FEATURES
eda_df['Timestamp'] = pd.to_datetime(eda_df['Timestamp'])
eda_df = eda_df.sort_values(['CustomerID', 'Timestamp'])




# Transactions in the Past 1 Hour
import numpy as np

# Convert timestamp
eda_df['Timestamp'] = pd.to_datetime(eda_df['Timestamp'], errors='coerce')
eda_df = eda_df.dropna(subset=['Timestamp'])

# Sort properly
eda_df = eda_df.sort_values(['CustomerID', 'Timestamp']).reset_index(drop=True)

# Empty column
eda_df['txns_last_1h'] = 0

# Loop per customer (FAST because each customer has few transactions)
for cid, group in eda_df.groupby('CustomerID'):
    times = group['Timestamp'].values
    result = np.zeros(len(times), dtype=int)
    
    for i in range(len(times)):
        # find how many timestamps fall within [current_time - 1h, current_time]
        cutoff = times[i] - np.timedelta64(1, 'h')
        result[i] = np.sum((times >= cutoff) & (times <= times[i]))
    
    eda_df.loc[group.index, 'txns_last_1h'] = result



# Transactions in the Past 24 Hours
import numpy as np

# Make sure timestamps are clean
eda_df['Timestamp'] = pd.to_datetime(eda_df['Timestamp'], errors='coerce')
eda_df = eda_df.dropna(subset=['Timestamp'])

# Sort properly
eda_df = eda_df.sort_values(['CustomerID', 'Timestamp']).reset_index(drop=True)

# Create empty columns
eda_df['txns_last_24h'] = 0
eda_df['amount_last_24h'] = 0.0

# Loop per customer (FAST because each customer has few transactions)
for cid, group in eda_df.groupby('CustomerID'):
    times = group['Timestamp'].values
    amounts = group['TransactionAmount'].values

    txn_count = np.zeros(len(times), dtype=int)
    amount_sum = np.zeros(len(times), dtype=float)

    for i in range(len(times)):
        cutoff = times[i] - np.timedelta64(24, 'h')

        mask = (times >= cutoff) & (times <= times[i])

        txn_count[i] = np.sum(mask)
        amount_sum[i] = np.sum(amounts[mask])

    eda_df.loc[group.index, 'txns_last_24h'] = txn_count
    eda_df.loc[group.index, 'amount_last_24h'] = amount_sum

# Amount spent in the last 24 hours
import numpy as np

# Ensure timestamp is cleaned and sorted
eda_df['Timestamp'] = pd.to_datetime(eda_df['Timestamp'], errors='coerce')
eda_df = eda_df.dropna(subset=['Timestamp'])

eda_df = eda_df.sort_values(['CustomerID', 'Timestamp']).reset_index(drop=True)

# Create empty output column
eda_df['amount_last_24h'] = 0.0

# Loop per customer ID
for cid, group in eda_df.groupby('CustomerID'):
    times = group['Timestamp'].values
    amounts = group['TransactionAmount'].values

    amount_sum_24h = np.zeros(len(times), dtype=float)

    for i in range(len(times)):
        cutoff = times[i] - np.timedelta64(24, 'h')

        # Mask of transactions within [current_time - 24h, current_time]
        mask = (times >= cutoff) & (times <= times[i])

        amount_sum_24h[i] = np.sum(amounts[mask])

    # Assign back to main df
    eda_df.loc[group.index, 'amount_last_24h'] = amount_sum_24h
```

```python
## 4. MERCHANT-LEVEL RISK FEATURES

eda_df['merchant_txn_count'] = eda_df.groupby('MerchantID')['TransactionAmount'].transform('count')
eda_df['merchant_avg_amount'] = eda_df.groupby('MerchantID')['TransactionAmount'].transform('mean')
eda_df['amount_diff_from_merchant_avg'] = eda_df['TransactionAmount'] - eda_df['merchant_avg_amount']



```

```python
# 2. one-hot encoder on categorical data 

from sklearn.preprocessing import OneHotEncoder
import pandas as pd
from typing import List, Tuple


def one_hot_encode_train(train_df: pd.DataFrame,) -> Tuple[pd.DataFrame, OneHotEncoder]:
    """
    Fit OneHotEncoder on training data and transform it.
    """

    categorical_cols = ['Category']

    ohe = OneHotEncoder(
        sparse_output=False,
        handle_unknown="ignore"
    )

    # Fit + transform on TRAIN
    ohe_array = ohe.fit_transform(train_df[categorical_cols])

    ohe_cols = ohe.get_feature_names_out(categorical_cols)

    ohe_df = pd.DataFrame(
        ohe_array,
        columns=ohe_cols,
        index=train_df.index
    )

    # Drop original categorical columns
    train_df_encoded = pd.concat(
        [train_df.drop(columns=categorical_cols), ohe_df],
        axis=1
    )

    return train_df_encoded, ohe
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
from typing import List, Tuple


def one_hot_encode_train_test(
    X_train: pd.DataFrame,
    X_test: pd.DataFrame,
    categorical_cols: List[str]
) -> Tuple[pd.DataFrame, pd.DataFrame, OneHotEncoder]:
    """
    Fit OneHotEncoder on training data and apply the same transformation on test data.
    """

    ohe = OneHotEncoder(
        sparse_output=False,
        handle_unknown="ignore"
    )

    # Fit ONLY on train
    train_array = ohe.fit_transform(X_train[categorical_cols])

    # Transform ONLY on test
    test_array = ohe.transform(X_test[categorical_cols])

    ohe_cols = ohe.get_feature_names_out(categorical_cols)

    train_ohe_df = pd.DataFrame(
        train_array,
        columns=ohe_cols,
        index=X_train.index
    )

    test_ohe_df = pd.DataFrame(
        test_array,
        columns=ohe_cols,
        index=X_test.index
    )

    X_train_encoded = pd.concat(
        [X_train.drop(columns=categorical_cols), train_ohe_df],
        axis=1
    )

    X_test_encoded = pd.concat(
        [X_test.drop(columns=categorical_cols), test_ohe_df],
        axis=1
    )

    return X_train_encoded, X_test_encoded, ohe




usage
categorical_cols = ["Category"]

X_train_encoded, X_test_encoded, ohe = one_hot_encode_train_test(
    X_train,
    X_test,
    categorical_cols
)
```

# 4. Scaler function

```python
from sklearn.preprocessing import RobustScaler, MinMaxScaler
import pandas as pd
from typing import List, Tuple, Dict


def scale_train_test(
    X_train: pd.DataFrame,
    X_test: pd.DataFrame,
    robust_features: List[str],
    minmax_features: List[str]
) -> Tuple[pd.DataFrame, pd.DataFrame, Dict[str, object]]:
    """
    Fit scalers ONLY on training data and apply them to both
    training and test datasets.

    Returns:
        - Scaled X_train
        - Scaled X_test
        - Dict of fitted scalers (for persistence)
    """

    X_train_scaled = X_train.copy()
    X_test_scaled = X_test.copy()

    scalers = {}

    # -------------------- RobustScaler --------------------
    if robust_features:
        robust_scaler = RobustScaler()
        X_train_scaled[robust_features] = robust_scaler.fit_transform(
            X_train[robust_features]
        )
        X_test_scaled[robust_features] = robust_scaler.transform(
            X_test[robust_features]
        )
        scalers["robust_scaler"] = robust_scaler
        scalers["robust_features"] = robust_features

    # -------------------- MinMaxScaler --------------------
    if minmax_features:
        minmax_scaler = MinMaxScaler()
        X_train_scaled[minmax_features] = minmax_scaler.fit_transform(
            X_train[minmax_features]
        )
        X_test_scaled[minmax_features] = minmax_scaler.transform(
            X_test[minmax_features]
        )
        scalers["minmax_scaler"] = minmax_scaler
        scalers["minmax_features"] = minmax_features

    return X_train_scaled, X_test_scaled, scalers
```

```python
import pandas as pd
from imblearn.combine import SMOTEENN


def oversample_smoteenn(
    train_df: pd.DataFrame,
    target_column: str
) -> pd.DataFrame:
    """
    Apply SMOTE-ENN oversampling ONLY on training data
    and return a single resampled training DataFrame.
    """

    # Split features and target
    X_train = train_df.drop(columns=[target_column])
    y_train = train_df[target_column]

    # Apply SMOTE-ENN
    smote_enn = SMOTEENN(random_state=42)
    X_resampled, y_resampled = smote_enn.fit_resample(X_train, y_train)

    # Combine back into a single DataFrame
    train_resampled = pd.concat(
        [
            pd.DataFrame(X_resampled, columns=X_train.columns),
            pd.Series(y_resampled, name=target_column)
        ],
        axis=1
    )

    return train_resampled
```
